```python
import pandas as pd
from datetime import datetime

# Ensure 'ScanDateTime' is in datetime format
df_week_filtered['ScanDateTime'] = pd.to_datetime(df_week_filtered['ScanDateTime'])

# Extract the date part from 'ScanDateTime'
df_week_filtered['ScanDate'] = df_week_filtered['ScanDateTime'].dt.date

# Remove rows where 'ScanStatus' is 'CCP'
df_week_filtered = df_week_filtered[df_week_filtered['ScanStatus'] != 'CCP']

def process_day_data(day_group):
    # Remove duplicates within the group
    day_group = day_group.drop_duplicates()

    # Fill NaN values in 'Zone' with 'BRI'
    day_group['Zone'] = day_group['Zone'].fillna('BRI')

    # Clean 'LabelNo' by stripping spaces and ensuring it's a string
    day_group['LabelNo'] = day_group['LabelNo'].astype(str).str.strip()

    # Filter out rows where latitude or longitude is 0
    day_group = day_group[
        (day_group['Scan_Latitude'] != 0) & (day_group['Scan_Longitude'] != 0)
    ]

    # Sort by 'ScanDateTime' in ascending order
    day_group = day_group.sort_values(by='ScanDateTime', ascending=True)

    # Keep only the first scan for each 'LabelNo'
    day_group = day_group.drop_duplicates(subset=['LabelNo'], keep='first')

    # Assign 'SequenceInRoute' based on the row's position in the group
    day_group['SequenceInRoute'] = (
        day_group.groupby(['CourierNumber', day_group['ScanDateTime'].dt.date])['ScanDateTime']
        .rank(method='first')
        .astype(int)
    )

    # Sort the data by 'CourierNumber' and 'ScanDateTime'
    day_group = day_group.sort_values(by=['CourierNumber', 'ScanDateTime'])

    # Return the processed dataframe for this day with all original columns
    return day_group

# Create a list to store processed day datasets
daily_datasets = []

# Process each days data and filter out invalid locations
for day, group in df_week_filtered.groupby('ScanDate'):
    print(f"Processing data for {day}: {group.shape[0]} rows")

    # Process the day's data
    processed_day_data = process_day_data(group)

    # Debug: Print row count for the processed day
    print(f"Processed {day}: {processed_day_data.shape[0]} rows")

    # Append the processed data to the list
    daily_datasets.append(processed_day_data)

# Combine all daily datasets into a single dataframe
combined_df = pd.concat(daily_datasets, ignore_index=True)

# Debug: Print total row count and the first few rows of the combined dataframe
print(f"Total rows after combining: {combined_df.shape[0]}")
combined_df

```

```python
# Create a copy of the original combined_df to perform operations
combined_df_copy = combined_df.copy()

# Ensure the ScanDate column is in datetime format
combined_df_copy['ScanDate'] = pd.to_datetime(combined_df_copy['ScanDate'])

# Group by address (DelAddress1, DelAddress2, DelTown, DelPostcode) and week
combined_df_copy['Week'] = combined_df_copy['ScanDate'].dt.isocalendar().week
combined_df_copy['Address'] = (
    combined_df_copy['DelAddress1'] + ' ' +
    combined_df_copy['DelTown'] + ' ' +
    combined_df_copy['DelPostcode'].astype(str)
)

# Identify consecutive day deliveries to the same address within the same week
combined_df_copy = combined_df_copy.sort_values(by=['Address', 'ScanDate'])
combined_df_copy['NextDeliveryDate'] = combined_df_copy.groupby('Address')['ScanDate'].shift(-1)

# Check if the next delivery is on the next day
combined_df_copy['ConsecutiveDay'] = (combined_df_copy['NextDeliveryDate'] - combined_df_copy['ScanDate']).dt.days == 1

# Count the number of unique addresses before and after bundling
initial_unique_stops_copy = combined_df_copy[['Address', 'ScanDate']].drop_duplicates().shape[0]

# Exclude today's stop if it can be bundled with tomorrow's
bundled_df_copy = combined_df_copy[~(combined_df_copy['ConsecutiveDay'])]

final_unique_stops_copy = bundled_df_copy[['Address', 'ScanDate']].drop_duplicates().shape[0]

# Calculate the reduction in total stops
reduction_in_stops_copy = initial_unique_stops_copy - final_unique_stops_copy

initial_unique_stops_copy, final_unique_stops_copy, reduction_in_stops_copy
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Start with the original dataframe (combined_df)
df_filtered = combined_df.copy()

# Step 2: Remove rows where 'ScanStatus' is 'PPP'
df_filtered = df_filtered[df_filtered['ScanStatus'] != 'PPP']

# Debug: Check total failed attempts (before further filtering)
total_failures_initial = df_filtered[df_filtered['Success/Failed'] == 'Failed'].shape[0]
print(f"Total initial failed deliveries: {total_failures_initial}")

# Step 3: Find all duplicated 'LabelNo' entries
duplicate_label_df = df_filtered[df_filtered.duplicated(subset='LabelNo', keep=False)]

# Debug: Check the shape of duplicate entries to ensure valid data is retained
print(f"Total duplicate 'LabelNo' entries: {duplicate_label_df.shape[0]}")

# Step 4: Sort by 'LabelNo' and 'ScanDateTime' to ensure chronological order
duplicate_label_df_sorted = duplicate_label_df.sort_values(by=['LabelNo', 'ScanDateTime'])

# Step 5: Define a function to count 'Failed' attempts before the first 'Success'
def count_failed_before_success(group):
    failed_count = 0
    success_found = False

    for _, row in group.iterrows():
        if row['Success/Failed'] == 'Failed' and not success_found:
            failed_count += 1
        elif row['Success/Failed'] == 'Success':
            success_found = True  # Stop counting after the first success

    return pd.Series({
        'LabelNo': group['LabelNo'].iloc[0],
        'Failed_Before_Success': failed_count
    })

# Step 6: Apply the function to each group of duplicated 'LabelNo'
failed_before_success_df = (
    duplicate_label_df_sorted.groupby('LabelNo')
    .apply(count_failed_before_success)
    .reset_index(drop=True)
)

# Debug: Check the total failed counts after processing
total_failed_count = failed_before_success_df['Failed_Before_Success'].sum()
print(f"Total failed attempts (after grouping): {total_failed_count}")

# Step 7: Calculate the frequency distribution
failed_attempts_distribution = (
    failed_before_success_df['Failed_Before_Success']
    .value_counts()
    .sort_index()
)

# Debug: Check if distribution adds up to the expected total
print(f"Failed attempts distribution:\n{failed_attempts_distribution}")

# Step 8: Plot the distribution
plt.figure(figsize=(10, 8))
bars = plt.bar(
    failed_attempts_distribution.index,
    failed_attempts_distribution.values,
    color='skyblue'
)

# Step 9: Add labels and title
plt.xlabel('Number of Failed Attempts Before Success', fontsize=12)
plt.ylabel('Number of Failed Deliveries', fontsize=12)
plt.title('Distribution of Failed Attempts Before Success', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Step 10: Annotate each bar with the raw count and percentage
for bar in bars:
    height = bar.get_height()
    percentage = (height / total_failed_count) * 100  # Calculate percentage
    plt.annotate(
        f'{int(height)} ({percentage:.1f}%)',
        xy=(bar.get_x() + bar.get_width() / 2, height),
        xytext=(0, 5),
        textcoords='offset points',
        ha='center', va='bottom'
    )

# Step 11: Ensure only integer ticks are displayed on the x-axis
max_tick = int(failed_attempts_distribution.index.max())
plt.xticks(ticks=np.arange(0, max_tick + 1, 1))

# Step 12: Adjust layout and display the plot
plt.tight_layout()
plt.grid(False)
plt.show()
```

```python
import pandas as pd
from datetime import datetime

# Ensure 'ScanDateTime' is in datetime format
df_week_filtered = combined_df.copy()
df_week_filtered['ScanDateTime'] = pd.to_datetime(df_week_filtered['ScanDateTime'])

# Extract the date part from 'ScanDateTime'
df_week_filtered['ScanDate'] = df_week_filtered['ScanDateTime'].dt.date

# Filter for data on July 1st
july_1st_data = df_week_filtered[df_week_filtered['ScanDate'] == datetime(2024, 7, 1).date()]

# Sort by 'LabelNo' and 'ScanDateTime'
july_1st_data = july_1st_data.sort_values(by=['LabelNo', 'ScanDateTime'])

# Remove rows with 'CCP' or 'PPP' statuses
july_1st_data = july_1st_data[~july_1st_data['ScanStatus'].isin(['CCP', 'PPP'])]

july_1st_data = july_1st_data[~july_1st_data['Success/Failed'].isin(['Success'])]

# Function to process duplicate and failed attempts in the required format
def process_failed_attempts(day_group):
    # Remove duplicates within the group
    day_group = day_group.drop_duplicates()

    # Fill NaN values in 'Zone' with 'BRI'
    day_group['Zone'] = day_group['Zone'].fillna('BRI')

    # Clean the 'LabelNo' column
    day_group['LabelNo'] = day_group['LabelNo'].astype(str).str.strip()

    # Sort by 'ScanDateTime'
    day_group = day_group.sort_values(by='ScanDateTime', ascending=True)

    # Create a new DataFrame with the required columns
    df_failed_attempts = pd.DataFrame({
        'Id': day_group['LabelNo'],
        'CustomerName': day_group['ScanStatus'] + '-' + day_group['Zone'],
        'AddressLine1': day_group['DelAddress1'],
        'AddressSuburb': day_group['DelTown'],
        'AddressState': None,  # No data provided
        'AddressPostcode': day_group['DelPostcode'],
        'Latitude': day_group['Scan_Latitude'],
        'Longitude': day_group['Scan_Longitude'],
        'ClientDeliveryWindowStart': '8:00',  # Fixed value
        'ClientDeliveryWindowFinish': '17:00',  # Fixed value
        'ClientServiceTime': 120,  # Fixed value in minutes
        'Zone': 'BRI',  # Default value
        'Quantity1': day_group['ParcelWeight'] * 1_000_000,  # Convert weight
        'RouteName': day_group['CourierNumber'],
        'SequenceInRoute': None,  # To be filled
        'MandatoryVehicles': day_group['CourierNumber'],
        'ScanDateTime': day_group['ScanDateTime']
    })

    # Assign 'SequenceInRoute' based on the rows position
    df_failed_attempts['SequenceInRoute'] = (
        df_failed_attempts.groupby(['RouteName', df_failed_attempts['ScanDateTime'].dt.date])['ScanDateTime']
        .rank(method='first').astype(int)
    )

    # Sort the DataFrame
    df_failed_attempts = df_failed_attempts.sort_values(by=['RouteName', 'ScanDateTime'])

    return df_failed_attempts

# Apply the processing function to the duplicate entries
processed_duplicates = july_1st_data.groupby('LabelNo', group_keys=False).apply(process_failed_attempts)

# Display the result
print(f"Total failed attempts for July 1st (including duplicates): {processed_duplicates.shape[0]}")
print(processed_duplicates)

# Optionally save the result to a CSV
#processed_duplicates.to_csv('failed_attempts_july_1st.csv', index=False)
```
